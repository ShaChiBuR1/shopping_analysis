```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('shopping_behavior.csv')
df.head()
df.tail()
df.info()
df.dtypes
df.isnull().sum()
df.duplicated().sum()
df.describe()
df.nunique()
df['Gender'].value_counts()
df['Category'].value_counts()
df['Payment Method'].value_counts()
df['Subscription Status'].value_counts()
numeric_df = df.select_dtypes(include=['int64', 'float64'])
numeric_df.corr()

plt.figure()
plt.hist(df['Age'], bins=20)
plt.title('Распределение возраста покупателей')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()

plt.figure()
plt.hist(df['Purchase Amount (USD)'], bins=20)
plt.title('Распределение суммы покупок')
plt.xlabel('Сумма покупки (USD)')
plt.ylabel('Количество')
plt.show()

df['Gender'].value_counts().plot(kind='bar')
plt.title('Распределение покупателей по полу')
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.show()

plt.figure()
plt.boxplot(df['Purchase Amount (USD)'])
plt.title('Boxplot суммы покупок')
plt.ylabel('USD')
plt.show()

plt.figure()
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Тепловая карта корреляций')
plt.show()

avg_purchase_by_gender = df.groupby('Gender')['Purchase Amount (USD)'].mean()
avg_purchase_by_gender
count_by_category = df.groupby('Category')['Customer ID'].count()
count_by_category
total_revenue_by_location = df.groupby('Location')['Purchase Amount (USD)'].sum()
total_revenue_by_location
stats_by_category = df.groupby('Category')['Purchase Amount (USD)'].agg(
    ['mean', 'median', 'sum', 'count']
)
stats_by_category
pivot_table = pd.pivot_table(
    df,
    values='Purchase Amount (USD)',
    index='Category',
    columns='Gender',
    aggfunc='mean'
)

pivot_table


```
